```python

import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
# Targets
y_gender = df_quan_scaled_merged['Sex_F']
y_adhd = df_quan_scaled_merged['ADHD_Outcome']

# Step 2: Fit MLR Model (for Gender prediction as example)
X = sm.add_constant(df_quan_scaled_merged[selected_features])
model = sm.OLS(y_gender, X).fit()

print(model.summary().as_text())
```

```python



# Fit model
X = sm.add_constant(df_quan_scaled_merged[selected_features])
y = df_quan_scaled_merged['Sex_F']  # or 'ADHD_Outcome'
model = sm.OLS(y, X).fit()

# Plot residuals vs fitted
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True)
plt.title("Residuals vs Fitted (Linearity Check)")
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
```

```python
from statsmodels.stats.stattools import durbin_watson

## 3.2 Independence
dw_stat = durbin_watson(model.resid)
print(f"Durbin-Watson statistic: {dw_stat:.4f}")
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan

bp_test = het_breuschpagan(model.resid, model.model.exog)
print("Breusch-Pagan p-value:", bp_test[1])  # small p â†’ heteroscedasticity
```

```python
import scipy.stats as stats

# Q-Q plot
sm.qqplot(model.resid, line='45')
plt.title("Q-Q Plot of Residuals")

# Shapiro-Wilk test
shapiro_test = stats.shapiro(model.resid)
print("Shapiro-Wilk p-value:", shapiro_test.pvalue)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

vif_data = pd.DataFrame()
vif_data['feature'] = selected_features
vif_data['VIF'] = [variance_inflation_factor(X.values, i+1) for i in range(len(selected_features))]
print(vif_data)
```

#### Most Predictive Features for Gender & ADHD


```python
#ðŸ’¡ Model: Mutual Information (nonlinear relationships)

from sklearn.feature_selection import mutual_info_classif

X = df_quan_scaled_merged[selected_features]
y_gender = df_quan_scaled_merged['Sex_F']
y_adhd = df_quan_scaled_merged['ADHD_Outcome']

mi_gender = mutual_info_classif(X, y_gender, discrete_features=False)
mi_adhd = mutual_info_classif(X, y_adhd, discrete_features=False)

# Create ranked tables
pd.Series(mi_gender, index=selected_features).sort_values(ascending=False)
pd.Series(mi_adhd, index=selected_features).sort_values(ascending=False)
```

```python
# ðŸ’¡ Model: Recursive Feature Elimination (RFE)
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE

rfe_model = RFE(LogisticRegression(), n_features_to_select=5)
rfe_model.fit(X, y_gender)  # or y_adhd
print("Top predictors:", [f for f, s in zip(selected_features, rfe_model.support_) if s])
```

```python
from sklearn.linear_model import LassoCV

lasso = LassoCV(cv=5).fit(X, y_gender)  # or y_adhd
lasso_coef = pd.Series(lasso.coef_, index=selected_features)
print("LASSO-selected features:", lasso_coef[lasso_coef != 0])
```
